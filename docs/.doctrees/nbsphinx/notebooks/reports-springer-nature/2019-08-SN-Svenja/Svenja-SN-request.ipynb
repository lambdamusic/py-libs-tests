````python
import dimcli
import os
import time
import pandas as pd
dsl = dimcli.Dsl()
````

````python
df = pd.read_csv("gridslist.csv")
````

````python
grids = df[['GRID ID','BPID']]
````

````python
q1 = """search publications where research_orgs="{}" and type="chapter" return publications"""
q2 = """search publications where research_orgs="{}" and type="chapter" return FOR limit 5"""
q3 = """search grants where research_orgs="{}" and active_year=2019 return grants"""
````

````python
# pull all data per a gridid, one call every 2 seconds
def pull_data_pubs(gridid):
    try:
        tot = dsl.query(q1.format(gridid)).stats['total_count']
    except:
        tot = 0
    time.sleep(1)
    return tot
    
def pull_data_grants(gridid):
    try:
        tot = dsl.query(q3.format(gridid)).stats['total_count']
    except:
        tot = 0
    time.sleep(1)
    return tot

def pull_data_FOR_pubs(gridid):
    try:
        res = dsl.query(q2.format(gridid))
        data = []
        for x in res.FOR[:5]:
            data.append(x['name'])
            data.append(x['count'])
    except:
        data = [""] * 10
    time.sleep(1)
    return data


def store_data(rows):
    labels = ['gridid', 'BPID', 'chapters', "FOR1", "FOR1count", "FOR2", "FOR2count", "FOR3", "FOR3count", "FOR4", "FOR4count", "FOR5", "FOR5count",  'grants2019', ]
    df = pd.DataFrame.from_records(rows, columns=labels)
    df.to_csv("output/final_data.csv")
````

````python
out = []
for index, row in grids.iterrows():
    gridid, BPID = row['GRID ID'], row['BPID']
    print(gridid)
    pubs = pull_data_pubs(gridid)
    grants = pull_data_grants(gridid)
    fors = pull_data_FOR_pubs(gridid)
    out.append((BPID, gridid, pubs, *fors, grants))
    store_data(out)
````
