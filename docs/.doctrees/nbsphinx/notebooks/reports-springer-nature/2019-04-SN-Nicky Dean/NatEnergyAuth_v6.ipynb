## Get researcher lists for journals

```python
# Define the SN Insights API search function that returns publications. 
# Journal ID needs to be fed in to it.  Returns publications by id, 
# publication year, affiliations, times cited, and Altmetric score.

def searchPubs(journal, year='>=2013', limit=1000, skip=0):
    data = """search publications
          where journal.id={} and year{} 
          return publications[
              id 
              + year 
              + author_affiliations 
              + times_cited 
              + title 
              + altmetric
              + doi
              + journal
              + concepts]
          limit {} skip {}
    """.format(journal,year,limit,skip)
    return data

# Define script to run complete search over all possible outcomes, 
# hence need for skip.

def dslsearchpublications(journal,year='>=2013'):
    skip = 0
    pubs = []
    total_pubs = []
    result = {}
    while (skip == 0) or (len(pubs) == 1000):
        pubs = dsl.query(searchPubs(journal=journal,year=year,skip=skip),0)['publications']
        total_pubs += pubs
        skip += 1000      
                            
    return total_pubs
```

```python
# Function to get full researchers but using dataframes instead of list
# comprehension, as originally.  Faster for bigger datasets.

def get_full_researchers(pubs):
        
    pubs = pubs.author_affiliations.apply(pd.Series).iloc[: ,0] \
            .apply(pd.Series) \
            .merge(pubs, left_index = True, right_index = True) \
            .drop(columns='author_affiliations') \
            .melt(id_vars = ['title',
                             'id',
                             'times_cited',
                             'altmetric',
                             'year',
                             'doi',
                             'journal',
                             'concepts'],
                  value_name = "author_affiliation") \
            .drop(columns="variable") \
            .dropna(subset=['author_affiliation'])

    pubs['name'] = ['{} {}'.format(d['first_name'], d['last_name'])
                       for d in pubs.author_affiliation]

    pubs['affil'] = [d['affiliations'] for d in pubs.author_affiliation]

    pubs['researcher_id'] = [d['researcher_id'] for d in pubs.author_affiliation]

    pubs = pubs.drop(columns='author_affiliation')[pubs['researcher_id'] != '']

    pubs = pubs.affil.apply(pd.Series) \
            .merge(pubs, left_index = True, right_index = True) \
            .drop(columns='affil') \
            .melt(id_vars = ['title',
                             'id',
                             'name',
                             'researcher_id',
                             'times_cited',
                             'altmetric',
                             'year',
                             'doi',
                             'journal',
                             'concepts'],
                  value_name = "author_affiliation") \
            .drop(columns="variable") \
            .dropna(subset=['author_affiliation'])#.sort_values('name')

    pubs['institution'] = [d['name'] for d in pubs.author_affiliation]
    pubs['city'] = [d.get('city') for d in pubs.author_affiliation]
    pubs['country'] = [d.get('country') for d in pubs.author_affiliation]
    pubs['grid'] = [d.get('id') for d in pubs.author_affiliation]

    pubs = pubs.drop(columns='author_affiliation')
    
    pubs.rename(columns={'name':'author', 
                         'times_cited':'cites',
                         'id':'dim_id'}, inplace=True)
    
    return pubs
```

## Get funding information for authors

```python
# API query for grants awarded since 2013 for researchers in the 
# researcherids list.  Returns the id, title, start/end dates, linkout, 
# funders and their country, researchers, their organisations, funding
# amount and resulting publications.

def grantsfromresearchers(researcherids,year=2013,limit=1000,skip=0):
    searchstring = """
    search grants
       where start_year>={} and researchers.id in [{}]
    return grants[id 
        + title 
        + start_year 
        + end_date 
        + linkout 
        + funders
        + funder_countries 
        + researchers 
        + research_org_name 
        + funding_usd
        + resulting_publication_ids 
        + researcher_details 
        + research_orgs]
        limit {} skip {}
    """.format(year,",".join([ '"{}"'.format(r) for r in researcherids]),limit,skip)
#     print (searchstring)
    return searchstring

def dslsearchgrantsR(ids,year=2013):
    skip = 0
    grants = []
    total_grants = []
    result = {}
    while (skip == 0) or (len(grants) == 1000):
        grants = dsl.query(grantsfromresearchers(ids,year,skip=skip),0)['grants']
        total_grants += grants
        skip += 1000                             
    return total_grants

def grantsfromresearcherlist(researcherids,year=2013):
    grants=[]
    
    idchunks = [researcherids[x:x + 500] for x in range(0, len(researcherids), 500)]
    for ids in idchunks:
        results = dslsearchgrantsR(ids,year)
        for x in results:
            if x not in grants:
                grants += [x]
#         print(len(grants))
        
    return grants
```

```python
def get_grantee_institutes(resgrants):
    grantee_inst = []

    for g in resgrants:
        dic = {}
        
        if 'research_orgs' in g.keys():
            dic['PI_institute_name'] = g.get('research_orgs')[0].get('name')
            dic['PI_institute_grid'] = g.get('research_orgs')[0].get('id')
        else:
            grant_inst = 'Unknown'

        dic['id'] = g.get('id')
        dic['Funding'] = g.get('funding_usd')
        dic['Start_year'] = g.get('start_year')
        dic['End_date'] = g.get('end_date')
        dic['Title'] = g.get('title')
        grantee_inst.append(dic)
    
    grantee_inst_df = pd.DataFrame(grantee_inst)
    grantee_inst_df['End_date'] = pd.to_datetime(grantee_inst_df['End_date'])
    grantee_inst_df['End_year'] = grantee_inst_df['End_date'].dt.year 
    grantee_inst_df['End_month'] = grantee_inst_df['End_date'].dt.month
    grantee_inst_df['End_2019'] = grantee_inst_df['End_year'] == 2019
    
    return grantee_inst_df
```

## Compile author lists for journals

```python
def get_authors(full_researchers, grantees):
    # Merge dataframes to build a dataframe that contains authors, their 
    # most recent affiliations, their number of publications, the sum/mean
    # of their citations and Altmetric scores, and their funding details.
    
    # First remove duplicates based on researcher ids and titles. This 
    # should produce the set of papers with just single authors per paper, 
    # ignoring multiple affiliations on each publication.
    no_dupes = full_researchers \
                .sort_values(['researcher_id','grid']) \
                .drop_duplicates(['researcher_id', 'title'])

    # Calculate the median cites and altmetrics of each researcher_id.
    median = no_dupes[['researcher_id','cites','altmetric']] \
                .groupby('researcher_id') \
                .median() \
                .rename(columns={'cites':'median_cites',
                                 'altmetric': 'median_altmetric'})

    # Calculate the count of publications for each researcher_id.
    count = no_dupes[['researcher_id','title']].groupby('researcher_id').count()

    # Get the sum grantee funding.
    grantees_sum = grantees[['id','Funding','End_2019']]. \
                    groupby('id').sum()

    # Combine everything
    auths = pd.concat([median, 
                       no_dupes.drop(columns=['cites', 'title', 'altmetric', 
                                              'doi', 'dim_id', 'concepts',
                                              'journal']) \
                                .drop_duplicates('researcher_id') \
                                .set_index('researcher_id'),
                       count,
                       grantees_sum
                       ], axis=1, sort=False) \
                .reset_index() \
                .rename(columns={'title':'publications',
                                 'index':'researcher_id',
                                 'year':'last_pub_year'}) \
                .dropna(subset=['author']) \
                .sort_values('Funding', ascending=False)
    
    # Replace researcher_ids with links.
    link = "https://sn-insights.dimensions.ai/discover/publication?and_facet_researcher="
    auths['researcher_id'] = (link + auths['researcher_id'])

    # Add columns to indicate if an entry is in the top 25% of each metric
    auths['publications_top_25pc'] = auths['publications'].rank(pct=True) >= 0.75
    auths['median_cites_top_25pc'] = auths['median_cites'].rank(pct=True) >= 0.75
    auths['median_altmetric_top_25pc'] = auths['median_altmetric'].rank(pct=True) >= 0.75
    auths['Funding_rank'] = auths['Funding'].rank(pct=True) >= 0.75

    # Reorder the dataframe to something more intuitively readable.
    auths = auths[[
        'author',
        'researcher_id',
        'institution',
        'grid',
        'city',
        'country',
        'publications',
        'median_cites',
        'median_altmetric',
        'Funding',
        'last_pub_year',
        'publications_top_25pc',
        'median_cites_top_25pc',
        'median_altmetric_top_25pc',
        'Funding_rank',
        'End_2019'
    ]].sort_values('publications', ascending=False).reset_index(drop=True)
    
    return auths
```

## Compile institutional information for journal authors

```python
# Build list of institutions and their papers.  Creates a list of 
# dictionaries; each dictionary corresponds to an affiliation for each 
# author.  For example, if there are two authors on a paper with two 
# affiliations each, there will be four dictionaries in the list.  As 
# such, there are many duplicates.  The aim is to get the publications 
# history of each institution, so we will need to remove these duplicates
# per paper.

def get_institutes_list(publications):
    # Initialise an empty list to be populated.
    inst = []

    # Cycle through the journal publications list.  Use if statements to
    # counteract instances where there is missing information for the 
    # affiliation.
    for g in publications:
        diclist = []
        if 'author_affiliations' in g.keys():
            for a in g['author_affiliations']:
                for affils in a:
                    for aff in affils['affiliations']:
                        dic = {}
                        dic['title'] = g['title']
                        dic['pub_id'] = g['id']
                        dic['year'] = g['year']
                        dic['cites'] = g['times_cited']
                        if 'altmetric' in g.keys():
                            dic['altmetric'] = g['altmetric']
                        dic['institution'] = aff['name']
                        if 'city' in aff.keys():
                            dic['city'] = aff['city']
                        if 'country' in aff.keys():
                            dic['country'] = aff['country']
                        if 'id' in aff.keys():
                            dic['grid'] = aff['id']
                        inst.append(dic)

    inst_df = pd.DataFrame(inst)
    return(inst_df)
```

```python
def get_institutes(publications, researchers, grantee_inst):

    # Define a function that will take in the dataframe of institutions and 
    # their publications, sort out the duplicates, calculate statistics and 
    # percentile indicators and then return that output.

    institutes_list = get_institutes_list(publications)
    
    # Get the list of unique institutions for all ERSS authors. This will 
    # provide base informationfor the merge below.
    institutions = institutes_list[['institution','grid','city','country']] \
                    .drop_duplicates(['institution'])
    
    # Merge dataframes to build a dataframe that contains institutions, 
    # their cities, countries, their number of publications in ERSS and  
    # the sum/mean of their citations (and Altmetric scores).

    # First remove duplicates based on institution and titles. This  
    # should produce the set of institutions for each paper.
    insts_no_dupes = institutes_list[['institution','title']] \
                        .drop_duplicates(['institution','title']) \
                        .groupby('institution') \
                        .count()

    # Get medians and sums of institutes and researchers.
    insts_median = institutes_list. \
                    drop_duplicates(['institution','title']) \
                    [['institution', 'cites', 'altmetric']] \
                    .groupby('institution').median() \
                    .rename(columns={'cites':'median_cites',
                                     'altmetric':'median_altmetric'})

    sum_researchers = researchers[['institution', 'researcher_id']] \
                        .drop_duplicates(['institution', 'researcher_id']) \
                        .groupby('institution').count() \
                        .rename(columns={'researcher_id':'authors_count'})

    grantee_inst_sum = grantee_inst[['PI_institute_name',
                                     'PI_institute_grid',
                                     'Funding', 
                                     'End_2019']] \
                        .groupby('PI_institute_name').sum()

    # Combine everything
    insts_list = pd.concat([insts_no_dupes,
                             insts_median,
                             institutions.set_index('institution'),
                             sum_researchers,
                             grantee_inst_sum],
                                axis=1, sort=False) \
                        .reset_index() \
                        .rename(columns={'title':'publications_count',
                                         'index':'institution'}) \
                        .dropna(subset=['authors_count'])

    # Replace researcher_ids with links.
    link = "https://sn-insights.dimensions.ai/discover/publication?and_facet_research_org=" 
    insts_list['grid'] = (link + insts_list['grid'])

    # Add columns to indicate if an entry is in the top 25% of each metric.
    insts_list['publications_count_top_25pc'] = insts_list['publications_count'].rank(pct=True) >= 0.75
    insts_list['authors_count_top_25pc'] = insts_list['authors_count'].rank(pct=True) >= 0.75
    insts_list['median_cites_top_25pc'] = insts_list['median_cites'].rank(pct=True) >= 0.75
    insts_list['median_altmetric_top_25pc'] = insts_list['median_altmetric'].rank(pct=True) >= 0.75
    insts_list['Funding_rank'] = insts_list['Funding'].rank(pct=True) >= 0.75
    
    # Reorder the dataframe to something more intuitively readable, then 
    # rename the columns more sensibly.
    insts_list = insts_list[[
        'institution',
        'grid',
        'city',
        'country',
        'publications_count',
        'authors_count',
        'Funding',
        'median_cites',
        'median_altmetric',
        'publications_count_top_25pc',
        'authors_count_top_25pc',
        'median_cites_top_25pc',
        'median_altmetric_top_25pc',
        'Funding_rank',
        'End_2019'
    ]].sort_values('publications_count', ascending=False).reset_index(drop=True)
    
    return insts_list  
```

```python
# Build a dictionary that contains all the journals ERSS authors have
# published in and their publication frequency.  Output a dataframe.

def journals_list(respubs):
    journals_list_auths = {}
    for p in respubs:
        if 'journal' in p: 
            if p['journal']['title'] not in journals_list_auths: 
                journals_list_auths[p['journal']['title']] = 1
            else:  
                journals_list_auths[p['journal']['title']] += 1
    # Make a dataframe of x, rename the columns.
    journals_list_auths_df = pd.DataFrame.from_dict(
        journals_list_auths,
        orient='index',
        columns=['count']
    )

    journals_list_auths_df = journals_list_auths_df.sort_values(
        'count', ascending=False
    ).reset_index().rename(columns={'index':'journal'})

    return journals_list_auths_df
```

## Define function to get data

```python
# Programmatic route to get everything.  Requires a journal name input 
# (look at journals variable in function for options) and the search year
# for the publications history search.  Note default grants search year is 2013.
# Output is a list of dataframes for the save function.

def get_journal_data(journal, researchers_year, grants_year, journals_year):
    
    journals = {
        'Nature_Energy': 'jour.1136447',
        'ERSS': 'jour.1053215',
        'Joule': 'jour.1298442',
        'Energy_Policy': 'jour.1120673',
        'EES': 'jour.1041992',
        'Adv_Energy_Mater': 'jour.1044941',
        'ACS_Energy_Letts': 'jour.1150266',
        'Nature_Physics': 'jour.1034717'
    }
    
    journal_id = journals.get(journal)
    # Run API script for publications.
    print("Getting the publications")
    pubs = dslsearchpublications(journal='"{}"'.format(journal_id), year=researchers_year)

    # Extract researcher ids from the journal's publications list.
    print("Getting the researcher ids")
    researchers = get_researcher_ids(pubs)

#     full_researchers = get_full_researchers(pubs, researchers)
    
    print("Getting the full researchers")
    full_researchers = get_full_researchers(pd.DataFrame(pubs))
    
    # Remove Joule editors from list for Joule
    if journal == 'Joule':
        full_researchers = full_researchers[
            full_researchers['author'] != 'Brandon R. Sutherland'
        ]
        full_researchers = full_researchers[
            full_researchers['author'] != 'Rahul Malik'
        ]
        full_researchers = full_researchers[
            full_researchers['author'] != 'Philip Earis'
        ]
    
    # Remove NE editors from list for NE
    if journal == 'Nature_Energy':
        full_researchers = full_researchers[
            full_researchers['author'] != 'Nature Energy'
        ]
        full_researchers = full_researchers[
            full_researchers['institution'] != 'Nature Energy, https://www.nature.com/natureenergy'
        ]
        full_researchers = full_researchers[
            full_researchers['institution'] != 'Nature Energy'
        ]
        
    # Get grants for authors of journal.
    print("Getting the grants")
    resgrants = grantsfromresearcherlist(list(set(researchers)), year=grants_year)

    # Get the information on grant awardees who are authors of the journal.
    print("Getting the grantees")
    grantees = get_grantees(pd.DataFrame(resgrants))

    # Get the information on institutions whose grant awardees who are 
    # authors of the journal.
    print("Getting the grantee institutions")
    grantee_inst = get_grantee_institutes(resgrants)

    # Get all the author information.
    print("Getting the authors")
    auths = get_authors(full_researchers, grantees)
    # auths_NE = auths_NE[auths_NE.institution != 'Nature Energy']

    # Get the information for institutions whose authors who publish in 
    # the journal.
    print("Getting the institutes list")
    insts_list = get_institutes(pubs, full_researchers, grantee_inst)

    # Get publications for journal authors.
    print("Getting the publications for researchers")
    respubs = publicationsfromresearcherlist(list(set(researchers)), year=journals_year)

    # Get the list of journals published in by authors of the journal.
    print("Getting the journals list")
    journals_list_auths = journals_list(respubs)
    
    years = {'researchers_year':researchers_year,
             'grants_year':grants_year,
             'journals_year':journals_year}
    
    # Get the list of concepts for each editor's content
    print("Getting the concepts list")
    termslist = pd.DataFrame(pubs).concepts.apply(pd.Series).melt() \
        .dropna().drop(columns='variable').value.value_counts().to_frame()
    
    # Original version of termslist:
    # termslist = full_researchers.concepts.apply(pd.Series).melt() \
    #    .dropna(subset=['value']).drop('variable', axis=1)
    # termslist_df = termslist['value'].value_counts().to_frame()
    # But this was counting among authors, so skewed terms to volume of 
    # authors.  Now doing it along just pubs, so should reflect content
    # not content x authors.
    
    journal_data = [auths, 
                    insts_list, 
                    journals_list_auths, 
                    years, 
                    full_researchers, 
                    termslist]
    print("Done")
    
    return journal_data
```

```python
# Note default grants search year is 2013.
# Output is a list of dataframes for the save function.

def get_editor_data(editor, grants_year, journals_year):
    
    NE_issue_contents = client.open('issue content reports')[0].get_as_df()
    NE_issue_contents = NE_issue_contents[NE_issue_contents['Article Type']. \
        isin(['Comment', 'News & Views','Perspective',
              'Review Article', 'Article', 'Analysis'])]
    
    # Include AR in FK data?
    eds = {
        'Alex': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'] == 'CZ'].DOI.tolist()],
        'James': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'] == 'JG'].DOI.tolist()],
        'Jenn': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'] == 'JR'].DOI.tolist()],
        'Nicky': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'] == 'ND'].DOI.tolist()],
        'Fouad': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'] == 'FK'].DOI.tolist()],
        'Alessandro': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'] == 'AR'].DOI.tolist()],
        'Giulia': [x.lower() for x in NE_issue_contents[
            NE_issue_contents['Editor'].isin(['GT', 'EC', 'ED'])].DOI.tolist()]
    }

    editor_doi_list = eds.get(editor)
    
    # Run API script for publications.
    pubs = publicationsfromdoislist(list(set(editor_doi_list)))
    
    # Extract researcher ids from the journal's publications list.
    researchers = get_researcher_ids(pubs)
    # Can this be replaced? Does it need its own function?
    
#     full_researchers = get_full_researchers(pubs, researchers)
    full_researchers = get_full_researchers(pd.DataFrame(pubs))
    
    # Get grants for authors of journal.
    resgrants = grantsfromresearcherlist(list(set(researchers)), year=grants_year)

    # Get the information on grant awardees who are authors of the journal.
#     grantees = get_grantees(resgrants)
    grantees = get_grantees(pd.DataFrame(resgrants))

    # Get the information on institutions whose grant awardees who are 
    # authors of the journal.
    grantee_inst = get_grantee_institutes(resgrants)

    # Get all the author information.
    auths = get_authors(full_researchers, grantees)
    # auths_NE = auths_NE[auths_NE.institution != 'Nature Energy']

    # Get the information for institutions whose authors who publish in 
    # the journal.
    insts_list = get_institutes(pubs, full_researchers, grantee_inst)

    # Get publications for journal authors.
    respubs = publicationsfromresearcherlist(list(set(researchers)), year=journals_year)

    # Get the list of journals published in by authors of the journal.
    journals_list_auths = journals_list(respubs)
    
    years = {'grants_year':grants_year,
             'journals_year':journals_year}
    
    # Get the list of terms for each editor's content
    termslist = pd.DataFrame(pubs).concepts.apply(pd.Series).melt() \
        .dropna().drop(columns='variable').value.value_counts().to_frame()
    
    # termslist = full_researchers.concepts.apply(pd.Series).melt().dropna(subset=['value']).drop(columns='variable')
    ## termslist.value.astype('category')
    # termslist_df = termslist['value'].value_counts().to_frame()
    
    journal_data = [auths, 
                    insts_list, 
                    journals_list_auths, 
                    years, 
                    full_researchers, 
                    pubs, 
                    termslist]
    
    return journal_data
```

```python
# API query for articles published with specific dois.
# Returns the id, year, author_affiliations and 
# journal of each publication.

def publicationsfromdois(doi_list,limit=1000,skip=0):
    searchstring = """
    search publications
       where doi in [{}]
    return publications[
        id 
        + year 
        + author_affiliations 
        + times_cited 
        + title 
        + altmetric
        + doi
        + journal
        + concepts] limit {} skip {}
    """.format(",".join([ '"{}"'.format(r) for r in doi_list]),limit,skip)
#     print (searchstring)
    return searchstring

# Define script to run complete search over all possible outcomes, 
# hence need for skip.

def dslsearchdois(ids):
    skip = 0
    pubs = []
    total_pubs = []
    result = {}
    while (skip == 0) or (len(pubs) == 1000):
        pubs = dsl.query(publicationsfromdois(ids,skip=skip),0)['publications']
        total_pubs += pubs
        skip += 1000      
                            
    return total_pubs

def publicationsfromdoislist(dois):
    pubs=[]
    
    idchunks = [dois[x:x + 400] for x in range(0, len(dois), 400)]
#     print(idchunks)
    for ids in idchunks:
        results = dslsearchdois(ids)
        for x in results:
            if x not in pubs:
                pubs += [x]
#         print(len(pubs))
        
    return pubs
```

## Define functions to look at who cites our content

```python
def citingpublications(ids,limit=1000,skip=0):
    data = """search publications
                  where (references in [{}]) and type="article"
                  return publications[id 
                                      + year 
                                      + author_affiliations 
                                      + times_cited 
                                      + title 
                                      + altmetric
                                      + doi
                                      + journal
                                      + concepts]
                  limit {} skip {}
        """.format(','.join(['"{}"'.format(id) for id in ids]),limit,skip)
    print(skip)
    return data

# Define script to run complete search over all possible outcomes, 
# hence need for skip.

def dslsearchpubids(ids):
    skip = 0
    pubs = []
    total_pubs = []
    result = {}
    while (skip == 0) or (len(pubs) == 1000):
        pubs = dsl.query(citingpublications(ids,skip=skip),0)['publications']
        total_pubs += pubs
        skip += 1000      
                            
    return total_pubs

def citingpubsfrompubidslist(pubids):
    pubs=[]
    
    idchunks = [pubids[x:x + 400] for x in range(0, len(pubids), 400)]
#     print(idchunks)
    for ids in idchunks:
        results = dslsearchpubids(ids)
        for x in results:
            if x not in pubs:
                pubs += [x]
#         print(len(pubs))
        
    return pubs
```

```python
def get_citing_data_filtered(journal_data, editor_data, grants_year,
                             filter_jnl, filter_self_cites=False):
    
    # journal_data is journal auths df, so should be NE_data[0].
    # editor_data is full_researchers df, so should be eg alex[4].
    # Two filters exist.  filter_jnl removes all NEnergy authors if true.
    # filter_self_cites removes the journal self-citing publications if
    # true. By default this is set to False.

    # Get publications citing Alex's publications list (this is a search query)...
    pubs = citingpubsfrompubidslist(editor_data.dim_id.to_list())
    pubs_df = pd.DataFrame(pubs)
    # ...then get the researcher ids and then full data for those papers.
    print("Getting the researcher ids...")
    citing_researchers = get_researcher_ids(pubs)
    print("...and the researchers' full data...")
    citing_full_researchers = get_full_researchers(pubs_df)
    print("Done")
    
    # Check filters.
    if filter_jnl == True:
        # From NE_data, extract the researcher_ids (could also output
        # researchers from get_journal_data function).
        jnl_researcher_ids = journal_data.researcher_id.str.replace(
            "https://sn-insights.dimensions.ai/discover/publication?and_facet_researcher=",
            "", regex=False).to_list()
        
        # Update citing_researcher variables if we are going to filter  
        # out the journal's authors.
        citing_researchers = list(set(citing_researchers) 
                                  - set(jnl_researcher_ids))
        citing_full_researchers = get_full_researchers(pubs_df)
        print("Will filter out journal authors")
        
    if filter_self_cites == True:
        # Remove journal self-citations (nsc - no self-citations).
        citing_full_researchers_nsc = citing_full_researchers[
            (~citing_full_researchers['doi'].str.contains('nenergy')) &
            (~citing_full_researchers['doi'].str.contains('s41560'))
            ]
        # This df contains no publications from NEnergy but it DOES  
        # contain NEnergy authors.
        
        # Create citing_researchers list that removes the self-cited info
        citing_researchers = citing_full_researchers.researcher_id.to_list()
        print("Will filter out journal self-citations")    

    # Get grants for authors who cite Alex's articles in the journal but 
    # aren't authors of the journal (runs a search query).
    print("Getting grants...")
    resgrants = grantsfromresearcherlist(citing_researchers, grants_year)

    # Get the information on grant awardees.
    print("Getting grant awardees...")
    grantees = get_grantees(pd.DataFrame(resgrants))

    # Get the information on institutions for the grant awardees.
    print("Getting grant awardee institutions...")
    grantee_inst = get_grantee_institutes(resgrants)

    # Get all the author information.
    print("Getting author information...")
    auths = get_authors(citing_full_researchers, grantees)

    # Get the information for institutions whose authors who publish in 
    # the journal.
    print("Getting institution information")
    insts_list = get_institutes(pubs, citing_full_researchers, grantee_inst)
    
    
    journals_list_citers = journals_list(pubs)
    
    
    years = {'grants_year':grants_year}
    
    # Get the list of terms for each editor's content    
    termslist = citing_full_researchers.concepts.apply(pd.Series) \
        .melt().dropna(subset=['value']).drop(columns='variable')
    # termslist.value.astype('category')
    termslist_df = termslist['value'].value_counts().to_frame()
    
    journal_data = [auths, insts_list, journals_list_citers, years, citing_full_researchers, pubs, termslist_df]
    
    return journal_data
```

```python
def relevant_papers(concepts, journal):
    
    df = pd.DataFrame(dsl.query(
            ("search publications where journal.id=\"{}\" in concepts for {} \
              return researchers[first_name + last_name] limit 1000") \
                .format(journal,
                        '"{}"'.format(" OR ".join(['{}'.format(c) for c in concepts]))
                       )).data['researchers'])

    df['author'] = df["first_name"] + ' ' + df["last_name"]
    
    return df[['author', 'id', 'count']]
```

```python
# Something is wrong with ACS Energy Letts concepts. Disabled.

def get_authors_from_concepts(concepts):
    journals = {
            'Nature Materials': 'jour.1031408',
            'Nature Nanotechnology': 'jour.1037429',
            'Nature Photonics': 'jour.1037430',
            'Nature Chemistry': 'jour.1041224',
            'Nature Reviews Materials': 'jour.1284373',
            'Nature Catalysis': 'jour.1299977',
            'Nature Reviews Chemistry': 'jour.1284555',
            'Advanced Materials': 'jour.1129018',
            'JACS': 'jour.1081898',
            'Angewandte': 'jour.1017044',
            'Joule': 'jour.1298442',
            'EES': 'jour.1041992',
            'Adv_Energy_Mater': 'jour.1044941',
#             'ACS_Energy_Letts': 'jour.1150266',
        }

    df_list = []
    author_list = []
    for j in journals:    
        df = relevant_papers(concepts, journals.get(j));

        author_list.append(pd.Series(df.author.values,index=df.id).to_dict())

        df.rename(columns={'count':j,
                           'author':'author {}'.format(j)},
                  inplace=True)

        df_list.append(df)

    authors = {}

    for a in author_list:
        authors.update(a)

    big_df = pd.concat([
        df.set_index('id') for df in df_list], axis=1, join='outer', sort=False
        ).reset_index()

    big_df['author'] = big_df.apply(
        lambda row: authors.get(row['index']),
        axis=1)

    cols = [c for c in big_df.columns if c.lower()[:7] != 'author ']
    cols = cols[-1:] + cols[:-1]
    big_df = big_df[cols].fillna(0)

    return(big_df)
```
