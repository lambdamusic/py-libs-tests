```python
keywords = """\\"electric machines\\" OR \\"electrical machines\\" OR \\"permanent magnet\\" OR brushless OR \\"PM machine\\" OR \\"switched reluctance\\" OR \\"synchronous reluctance\\" """
```

```python
pubs_agg_query = """search publications in title_abstract_only for """ + '"' + keywords + '"' + \
    """ where year in [2013:2019] and research_orgs.id = "grid.10025.36"
    return research_orgs aggregate rcr_avg, fcr_gavg, altmetric_median"""
```

```python
dsl = dimcli.Dsl(instance='test')
res = dsl.query(pubs_agg_query)
```
