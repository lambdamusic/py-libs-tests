# All journals for 2019

```python
import time
import pandas
import dimcli
dimcli.login()
```

A handy function to generate a list of day-dates from a date range:

```python
from datetime import datetime, date, timedelta
def date_range_from_strings(start_date=None, end_date=None):
    """
    From two string-dates formatted as YYYY-MM-DD
    Return a delta date_object which can be iterated on  
    <start_date> : => default: YESTERDAY
    <end_date> : => default: TODAY
    """
     # the following is redundant, added just as an example to simulate receiving a date-string
    if not start_date:
        yesterday = date.today() - timedelta(1)
        start_date = yesterday.strftime("%Y-%m-%d")
    if not end_date:
        end_date = date.today().strftime("%Y-%m-%d")
    d1 = datetime.strptime(start_date, '%Y-%m-%d').date()
    d2 = datetime.strptime(end_date, '%Y-%m-%d').date()
    delta = d2 - d1
    out = []
    for i in range(delta.days):
        step = d1 + timedelta(i)
        out += [step.strftime("%Y-%m-%d")]
    return out
# EG
start_date = "2019-01-01"
end_date = None # = today
print(date_range_from_strings(start_date, end_date))
```

```python
q_template = """search publications where date="%s" return journal limit 1000"""
dsl = dimcli.Dsl()
```

Now we iterate through all the day-dates available and pull out all journals from each date. The journals are stored in an index, and the loop prints out information as it go through all dates (note: one query every 2 seconds to comply with the API max requests limit). 

```python
journals = {}
start_date = "2019-01-01"
end_date = None # = today
for day in date_range_from_strings(start_date, end_date):
    print(f"Date: = {day}")
    res = dsl.query(q_template % day, show_results=False)
    if res.journal:
        for x in res.journal:
            try:
                journals[x['id']] = x['title']
            except:
                pass # journal already indexed
    print(f">>> journals in index: {len(journals)}")
    time.sleep(2)
```

```python
df = pandas.DataFrame([(x, journals[x]) for x in journals])
df.columns = ['id', 'title']
df
```
